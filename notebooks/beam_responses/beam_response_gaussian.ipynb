```python
def rad_interp(J_section, az, alt):
    """
    Abbreviation and conversion function. We get the interpolation value
    assuming @az and @alt are in degrees.
    """
    colatitude = np.radians(90 - alt)
    longitude = np.radians(az)
    return hp.get_interp_val(J_section, colatitude, longitude)
```

```python
# :: xy and yx are just leakage terms; they are not beam response terms
# in other words, the maximum response can occur at any point, and it is okay that these do not maximize.

print(rad_interp(J_abs[:, 1, 0], az, alt)) # true maximum: .998166
print(rad_interp(J_abs[:, 0, 0], az, alt)) # ": .0624716
print(rad_interp(J_abs[:, 0, 1], az, alt)) # ": .998166
print(rad_interp(J_abs[:, 1, 1], az, alt)) # ": .0624716
```

```python
def response_plot(label, section):
    fig, ax = sf.demo.frame() # this is somewhat abstruse;
        # maybe we should move the method to a different module. We should at least rename.
    plt.title(label + ' varying altitude about ' + str(round(alt, 1)) + '$^\circ$')
    ax.plot(alt_neigh, rad_interp(section, az_const, alt_neigh))
    plt.ylabel('Response [0, 1]', fontsize=12)
    plt.xlabel('Altitude [$^\circ$]')
    plt.show()
    
    fig, ax = sf.demo.frame()
    plt.title(label + ' varying azimuth about ' + str(round(az, 1)) + '$^\circ$')
    ax.plot(az_neigh, rad_interp(section, az_neigh, alt_const))
    plt.ylabel('Response [0, 1]', fontsize=12)
    # I have to force this y window if I want the axis to look reasonable.
    plt.ylim((0, 1.1))
    plt.xlabel('Azimuth [$^\circ$]')
    plt.show()
```

```python
def rotate_by(angle, radians=False):
    if not radians:
        angle = np.radians(angle)
    s = np.sin(angle)
    c = np.cos(angle)
    return np.array([
        [c, -s],
        [s, c]
    ])
```

```python
theta_space = np.linspace(0, 2 * np.pi, 100)
J_pred = []
for theta in theta_space:
    J_pred.append(np.dot(rotate_by(theta, radians = True), J[0]))
J_pred = np.abs(np.array(J_pred))
```

```python
# plain old direct rotation, as a baseline for comparison
plt.plot(theta_space, J_pred[:, 0, 0])
```

```python
J_dpred = []
for theta in theta_space:
    R = rotate_by(theta, radians=True)
    Ri = np.linalg.inv(R)
    J_dpred.append(np.dot(R, np.dot(J[0], Ri)))
    #print(J_dpred[len(J_dpred) - 1])
J_dpred = np.abs(np.array(J_dpred))
```

```python
# change of basis rotation
plt.plot(theta_space, J_dpred[:, 0, 0])
# It is simply too smooth
```

```python
ha_to_topo = sf.rot.M_ha_to_topo(radians = True)
print(ha_to_topo)

def hrotate_by(angle, radians=False):
    if not radians:
        angle = np.radians(angle)
    
    s = np.sin(angle)
    c = np.cos(angle)
    
    rectL = sf.rot.rectangle(c, s, radians=True)
    #print(rectL)
    topo_rotL = np.dot(ha_to_topo, rectL)
    #print(topo_rotL)
    outL = sf.rot.new_sphere(topo_rotL, radians=True)
    #print(outL)
    
    rectR = sf.rot.rectangle(-s, c, radians=True)
    topo_rotR = np.dot(ha_to_topo, rectR)
    outR = sf.rot.new_sphere(topo_rotR, radians=True)
    
    return np.vstack((outL, outR))
```

```python
J_hpred = []
for theta in theta_space:
    #print(hrotate_by(theta, radians = True))
    J_hpred.append(np.dot(hrotate_by(theta, radians = True), J[0]))
J_hpred = np.abs(np.array(J_hpred))
```

```python
# what if we convert the rotation matrix to topocentric coordinates?
plt.plot(theta_space, J_hpred[:, 0, 0])
plt.plot(theta_space, J_hpred[:, 0, 1])
plt.plot(theta_space, J_hpred[:, 1, 0])
plt.plot(theta_space, J_hpred[:, 1, 1])
```
